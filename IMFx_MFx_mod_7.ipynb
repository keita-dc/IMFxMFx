```python
quiz_3_df = pd.read_csv("data/M7_S2_quiz_3.csv")
```

```python
quiz_3_df.head()
```

```python
quiz_3_df.info()
```

```python
quiz_3_df["y_correct"] = 0.5* quiz_3_df.z + 0.4 * quiz_3_df.epsilon.shift(1) + 0.5 * quiz_3_df.epsilon.shift(2)
```

```python
model = sm.OLS(endog=quiz_3_df.y[:950], exog=quiz_3_df.z[:950]).fit()
model.summary()
```

```python
quiz_3_df["y_incorrect"] = model.predict(exog=quiz_3_df.z)
```

```python
print(np.sqrt(mean_squared_error(quiz_3_df.y[950:], quiz_3_df.y_correct[950:])))
print(np.sqrt(mean_squared_error(quiz_3_df.y[950:], quiz_3_df.y_incorrect[950:])))
```

```python
plt.figure(figsize=(18,6))
plt.plot(quiz_3_df.y[900:])
plt.plot(quiz_3_df.y_correct[900:])
plt.plot(quiz_3_df.y_incorrect[900:])
plt.legend(["actual", "correct", "incorrect"])
plt.show()
```
